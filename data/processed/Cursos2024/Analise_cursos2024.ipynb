## 2. Merge

```python
# Pega apenas o nome do curso de fato
df['CURSO'] = df['CURSO'].apply(lambda x: x.split('-')[0].strip())

# DropDuplicates se tiver
print(f'Shape before: {df.shape}')
df = df.drop_duplicates()
print(f'Shape after: {df.shape}')
```

```python
cols_pesquisa = df_pesquisa.columns
cols_cursos = df_cursos.columns

diff_cols_pesquisa_cursos = cols_cursos.difference(cols_pesquisa)
print(diff_cols_pesquisa_cursos)
```

```python
if len(diff_cols_pesquisa_cursos) > 0:
    print(f'Shape before 2° merge: {df_pesquisa.shape}')
    df_pesquisa = pd.merge(
        left=df_cursos[['CURSO','SETOR_CURSO']],
        right=df_pesquisa,
        on='CURSO',
        how='left'
    )
    print(f'Shape after 2° merge: {df_pesquisa.shape}')
    df_pesquisa = df_pesquisa.dropna()
    print(f'Shape after dropna: {df_pesquisa.shape}')

```

```python
df_merged_perguntas_dimensoes = pd.merge(
    left=df_info_perguntas.rename(columns={'Tipo_Pergunta':'Tipo_Perg'}),
    right=df_info_dimensoes[['Tipo_Perg','EIXO','DIMENSAO']],
    on='Tipo_Perg',
    how='left')
```

```python
print(df_merged_perguntas_dimensoes.columns.difference(df_pesquisa.columns))
```

```python
if 'Ordem' not in df_pesquisa.columns:
    print('Shape Before: ', df_pesquisa.shape)
    df_pesquisa = pd.merge(
        left=df_pesquisa,
        right=df_merged_perguntas_dimensoes[['ID_PERGUNTA','CL_PERGUNTA', 'DIMENSAO', 'EIXO', 'Ordem', 'Tipo_Perg']],
        on='ID_PERGUNTA',
        how='left')
    print('Shape After: ', df_pesquisa.shape)
```

## 3. Exportação

```python
def process(df: pd.DataFrame, axis: list[int] | None = None, dimensions: list[int] | None = None, questions: list[str] | None = None) -> pd.DataFrame:
    
        if axis is None or axis == []:
              axis = df['EIXO'].unique().tolist()
            
        if dimensions is None or dimensions == []:
            dimensions = df['DIMENSAO'].unique().tolist()

        mask_axis = df['EIXO'].isin(axis)
        mask_dim = df['DIMENSAO'].isin(dimensions)

        df_filtered = df[mask_axis & mask_dim]

        if questions is not None:
            df_filtered = df_filtered[df_filtered['PERGUNTA'].isin(questions)]

        return df_filtered
```

```python
df_final = process(df_pesquisa,axis=[],dimensions=[])
```

```python
ordem_atual = df_final.columns
print('Ordem Atual: ',ordem_atual.tolist())

nova_ordem = ['EIXO','DIMENSAO','ID_PESQUISA', 'ID_PERGUNTA', 'ID_QUESTIONARIO', 'QUESTIONARIO', 'PERGUNTA', 'RESPOSTA', 'SITUACAO', 'MULTIPLA_ESCOLHA', 'CURSO', 'COD_CURSO', 'SETOR_CURSO', 'CL_PERGUNTA', 'Ordem', 'Tipo_Perg']

df_final = df_final[nova_ordem]
```

```python
df_final['VALOR_RESPOSTA'] = [1 if i in ['Concordo','Sim'] else 0 if i == 'Desconheço' else -1 for i in df_final['RESPOSTA']]
```

```python
df_final.to_csv('processed_cursos_2024.csv',index=False)
```
